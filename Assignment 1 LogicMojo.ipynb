Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

#1. Identify and display all transaction values that appear more than once.
ans1 = []
for x in transactions : 
    c = transactions.count(x)
    if c > 1 and x not in ans1:
        ans1.append(x)

print("1. Identify and display all transaction values that appear more than once")
print(ans1)


print("\n")

#2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
ans2 = []
for x in transactions :
    if x > 5000 :
        ans2.append(x)

print("2. Detect transactions greater than ₹5000 as these may require fraud monitoring.")
print(ans2)

print("\n")

#Calculate the total spending for the day.

ans3 = sum(transactions)
print("3. Calculate the total spending for the day.")
print(ans3)

print("\n")


#4. Create a new list containing only unique transaction values.

ans4 = []

for x in transactions :
    if x not in ans4 :
        ans4.append(x)

print("4. Create a new list containing only unique transaction values.")
print(ans4)

print("\n")

#5. Sort the unique transaction list in descending order to view the largest transactions first.

ans5 = []
ans5 = transactions
ans5.sort(reverse=True)

print("5. Sort the unique transaction list in descending order to view the largest transactions first.")
print(ans5)
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]


# 1. Iterates through the dataset and removes all invalid age values.
invalid = []

print(" 1. Iterates through the dataset and removes all invalid age values.")
print("Ages before : ",ages)

for a in ages : 
    if a == None or a < 0 : 
        invalid.append(a)
        ages.remove(a)

print("Ages after  : ",ages)
print("\n")


# 2. Stores only valid ages in a new list.
valid = []
valid = ages

print("2. Stores only valid ages in a new list.")
print(valid)
print("\n")

# 3. Calculates the average age of valid customers.
avg = sum(valid)/len(valid)
print("3. Calculates the average age of valid customers.")
print(avg)

print("\n")

# 4. Displays the list of customers whose age is greater than 30.
ans = []
for a in valid : 
    if a > 30 : 
        ans.append(a)

print("4. Displays the list of customers whose age is greater than 30.")
print(ans)
print("\n")


# 5. Prints how many invalid entries were removed during the cleaning process.
print("5. Prints how many invalid entries were removed during the cleaning process.")
print(invalid)
```

Question 5:- An e-commerce company exports its daily order log in the following format.
Each order contains multiple pieces of information:

orders = [
["ORD001","Laptop",75000,"Completed"],
["ORD002","Mouse",500,"Completed"],
["ORD003","Laptop",75000,"Returned"],
["ORD004","Keyboard",1500,"Completed"],
["ORD005","Laptop",75000,"Completed"],
["ORD006","Monitor",20000,"Fraud"],
["ORD007","Mouse",500,"Completed"],
["ORD008","Laptop",75000,"Completed"]
]

Each row represents an order with the following structure:

[Order_ID , Product_Name , Price , Status]

Where:
Order_ID → unique order identifier
Product_Name → product purchased
Price → price of the product
Status → order status (Completed, Returned, Fraud)

Your Task:

Write a Python program to simulate a simple order analysis system using loops, nested loops, conditional statements, and list operations.

Your program must perform the following steps.

1. Display All Orders
Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly.

2. Detect Suspicious Activity
The system suspects fraud when:
the same product appears more than 3 times, OR
an order has status "Fraud"

Write a program that scans the dataset and identifies suspicious products or orders.
You must use nested loops to compare product occurrences.

3. Calculate Total Revenue
Only Completed orders contribute to revenue.
Write code that:
iterates through the dataset
adds the price of all completed orders
prints the total revenue generated for the day.

4. Count Order Types
Using loops and conditional statements, count how many orders fall into each category:

Completed
Returned
Fraud
Store these counts in separate variables.

5. Generate a List of Unique Products
Create a new list containing unique product names present in the dataset.
You must implement this without using the set() function.

6. Product Sales Analysis
For each unique product, determine:
how many times it was ordered
total revenue generated by that product
This requires nested loops.

7. Identify the Most Sold Product
Based on the frequency of orders, determine which product appears the most in the dataset.

8. Remove Fraud Orders
Create a cleaned dataset by removing orders whose status is "Fraud".
Use list methods to store the cleaned data in a new list.

```python
orders = [
["ORD001","Laptop",75000,"Completed"],
["ORD002","Mouse",500,"Completed"],
["ORD003","Laptop",75000,"Returned"],
["ORD004","Keyboard",1500,"Completed"],
["ORD005","Laptop",75000,"Completed"],
["ORD006","Monitor",20000,"Fraud"],
["ORD007","Mouse",500,"Completed"],
["ORD008","Laptop",75000,"Completed"]
]

# 1. Display All Orders
print("All Orders:\n")
for ord in orders : 
    for items in ord : 
        print(items,end=' ')
    print()
    

# 2. Detect Suspicious Activity
print("\nSuspicious Activity:")

# Check product repetition
for i in range(len(orders)):
    count = 0
    for j in range(len(orders)):
        if orders[i][1] == orders[j][1]:
            count += 1
    if count > 3:
        print("Product appearing too many times:", orders[i][1])
        break

# Check fraud orders
for order in orders:
    if order[3] == "Fraud":
        print("Fraud Order Found:", order)


# 3. Calculate Total Revenue
total_revenue = 0
for order in orders:
    if order[3] == "Completed":
        total_revenue += order[2]

print("\nTotal Revenue:", total_revenue)


# 4. Count Order Types
completed = 0
returned = 0
fraud = 0

for order in orders:
    if order[3] == "Completed":
        completed += 1
    elif order[3] == "Returned":
        returned += 1
    elif order[3] == "Fraud":
        fraud += 1

print("\nOrder Counts:")
print("Completed:", completed)
print("Returned:", returned)
print("Fraud:", fraud)


# 5. Unique Products (without set)
unique_products = []

for order in orders:
    if order[1] not in unique_products:
        unique_products.append(order[1])

print("\nUnique Products:", unique_products)


# 6. Product Sales Analysis
print("\nProduct Analysis:")

for product in unique_products:
    count = 0
    revenue = 0
    
    for order in orders:
        if order[1] == product:
            count += 1
            if order[3] == "Completed":
                revenue += order[2]
    
    print(product, "-> Orders:", count, ", Revenue:", revenue)


# 7. Most Sold Product
max_count = 0
most_sold = ""

for product in unique_products:
    count = 0
    for order in orders:
        if order[1] == product:
            count += 1
    
    if count > max_count:
        max_count = count
        most_sold = product

print("\nMost Sold Product:", most_sold)


# 8. Remove Fraud Orders
clean_orders = []

for order in orders:
    if order[3] != "Fraud":
        clean_orders.append(order)

print("\nCleaned Orders:")
for order in clean_orders:
    print(order)
```
