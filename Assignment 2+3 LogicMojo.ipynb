Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers.
Skip invalid values using exception handling.

````python
data = ["10", "20", "abc", "30"]
for i in range (0,len(data)-1,1) : 
    try : 
        data[i] = int(data[i])
    except ValueError :
        data.pop(i)
        print("Value at index ",i," is not an integer hence erasing it")

print(data)
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
import re

def clean_text(text):
    # 1. Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # 2. Convert to lowercase
    text = text.lower()
    
    # 3. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
cleaned_texts = []

for t in texts:
    cleaned_texts.append(clean_text(t))

print(cleaned_texts)
````

transactions = [

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
def detect_increasing_sequence(data):
    if not data:
        return []

    longest = []
    current = [data[0]]

    for i in range(1, len(data)):
        if data[i] > data[i - 1]:
            current.append(data[i])
        else:
            
            if len(current) > len(longest):
                longest = current
            current = [data[i]]

    if len(current) > len(longest):
        longest = current

    return longest

data = [10,20,30,50,30]


print(detect_increasing_sequence(data))

````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---







````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    if score < 50:
        return "Rejected"

    return decision_tree(score + 10)

print(decision_tree(454))  
print(decision_tree(600))  
print(decision_tree(802))  
````
